```python
# %%time
# train, test = train_test_split(df, test_size=0.2, random_state=42)
train, test = df.split_frame(ratios=[.8], seed = 42)
```

```python
# y = 'is_canceled'
y = 'Churn'
x = list(df.columns)
x.remove(y)
```

```python
from h2o.automl import H2OAutoML
```

```python
%%time
aml = H2OAutoML(seed = 42, exclude_algos = ["DeepLearning"], max_runtime_secs_per_model=30, verbosity="info")
```

```python
%%time
aml.train(x = x, y = y, training_frame = train)
```

```python
leaderboard = aml.leaderboard
leaderboard
```

# Prediction on test data

```python
data_pred=aml.leader.predict(test)
```
