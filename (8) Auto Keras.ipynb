```python
import numpy as np
import pandas as pd
import tensorflow as tf

import autokeras as ak
```

```python
from sklearn.model_selection import train_test_split
    
```

```python
df = pd.read_csv(r'C:\Users\himgoyal.ext\Desktop\MY DRIVE\AutoML\Experiment\Data\WA_Fn-UseC_-Telco-Customer-Churn.csv')
df
```

```python
X = df.drop(['Churn'], axis=1)
y = df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.80, random_state=42)
```

```python
%%time
# Initialize the structured data classifier.
clf = ak.StructuredDataClassifier(overwrite=True, max_trials=10, seed = 42)  # It tries 3 different models.
# Feed the structured data classifier with training data.
clf.fit(X_train, y_train, validation_split=0.2, epochs=10)
```

```python
print(type(clf))
model = clf.export_model()
model.summary()
```

```python
# Predict with the best model.
predicted_y = clf.predict(X_train)
```

```python
# Evaluate the best model with testing data.
print(clf.evaluate(X_test, y_test))
```
